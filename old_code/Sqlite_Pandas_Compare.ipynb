```python
#a) Calculate top 25 most common 'makes'
## Using Pandas Dataframe

df_common_make = df[['Ticket number','Make']].groupby(['Make']) \
                                                            .agg('count') \
                                                            .sort_values('Ticket number', ascending = False) \
                                                            .head(25)\
                                                            .reset_index()\
                                                            
l_common_make = df_common_make['Make'].tolist()

l_common_make
```

```python
# Execution Time Using Timeit
## for Pandas

import timeit

%timeit (df[['Ticket number','Make']].groupby(['Make']).agg('count').sort_values('Ticket number', ascending = False).head(25))
               
```

```python
# a)Calculate top 25 most common 'Makes'
## Using Sqlite


engine.execute("SELECT Make FROM parking_citations GROUP BY Make ORDER BY count('Ticket number') DESC LIMIT 25;").fetchall()
```

```python
#Question:
#"Is an out-of-state license plate more likely to be expired 
#at the time of receiving a ticket than an in-state license plate?" 

# Explory a comparison to identify if the vehicle had an expired plate at the ticket issue date
df = df.dropna(axis = 'index', subset = ['Plate Expiry Date'])
df[df['Issue Date'].astype(float) > df['Plate Expiry Date'].astype(float)*100]
```

```python
# Creating "Is Expired" Variable


df['Is Expired'] = df['Issue Date'].astype(float) > df['Plate Expiry Date'].astype(float)*100


# Creating "out-of-state" Variable
# Assuming CA as the State conducting the analysis


df['Out_of_State'] = [ 0 if x == 'CA' else 1 for x in df['RP State Plate']]
df
```

```python
#Question:
#"Is an out-of-state license plate more likely to be expired 
#at the time of receiving a ticket than an in-state license plate?" 

#Answer:
#YES


df.groupby('Out_of_State')['Is Expired'].agg(pd.Series.mean).to_frame().reset_index()
```

```python
df.groupby('Out_of_State')['Is Expired'].agg(pd.Series.sum).to_frame().reset_index()
```
